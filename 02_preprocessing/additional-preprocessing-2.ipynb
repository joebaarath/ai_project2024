```python
import pandas as pd
import contractions

# Function to expand contractions in a text
def expand_contractions(text):
    expanded_words = []
    for word in text.split():
        expanded_words.append(contractions.fix(word))
    return ' '.join(expanded_words)

# Read data from the database (assuming it's in a DataFrame format)
df = pd.read_csv('/kaggle/working/test_combined.csv')
# Apply contractions expansion to the specified columns
for col in ['Headline', 'articleBody']:  # Replace 'column1' and 'column2' with actual column names
    df[col] = df[col].apply(expand_contractions)

# Save the DataFrame to a new CSV file
df.to_csv('expanded_test.csv', index=False)

print("Expansion complete. Data saved to expanded_database.csv")
```

```python
import pandas as pd
import re
def convert_to_lower(text):
    return text.lower()

# Function to remove URLs
def remove_urls(text):
    return re.sub(r'http\S+', '', text)

# Function to remove special symbols
def remove_special_symbols(text):
    return re.sub(r'[^\w\s]', '', text)

# Apply the defined functions to the columns
df['Headline'] = df['Headline'].apply(convert_to_lower)
df['Headline'] = df['Headline'].apply(remove_urls)
df['Headline'] = df['Headline'].apply(remove_special_symbols)

df['articleBody'] = df['articleBody'].apply(convert_to_lower)
df['articleBody'] = df['articleBody'].apply(remove_urls)
df['articleBody'] = df['articleBody'].apply(remove_special_symbols)
```

```python
import pandas as pd
import re
import string
def preprocess_text(text):
    text = re.sub(r'%', 'percent', text)
    # Replace '&' with 'and'
    text = re.sub(r'&', 'and', text)
    # Remove symbols, including punctuation marks and special characters
    text = re.sub(r'[^a-zA-Z0-9\s]', ' ', text)
    # Remove Chinese characters
    text = re.sub(r'[\u4E00-\u9FFF]+', ' ', text)
    # Remove URLs
    text = re.sub(r'http\S+|www\S+', ' ', text)
    # Remove mentions
    text = re.sub(r'@\w+', ' ', text)
    # Remove emojis
    emoj = re.compile("["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
        u"\U00002500-\U00002BEF"  # Chinese char
        u"\U00002702-\U000027B0"
        u"\U000024C2-\U0001F251"
        u"\U0001f926-\U0001f937"
        u"\U00010000-\U0010ffff"
        u"\u2640-\u2642"
        u"\u2600-\u2B55"
        u"\u200d"
        u"\u23cf"
        u"\u23e9"
        u"\u231a"
        u"\ufe0f"
        u"\u3030"
                      "]+", re.UNICODE)
    text = re.sub(emoj, ' ', text)
    # Remove punctuation
    text = re.sub(r'[%s]' % re.escape(string.punctuation), ' ', text)
    # Remove non-alphanumeric characters
    text = re.sub(r'[^a-zA-Z0-9]', ' ', text)
    text = text.lower()
    return text

# Preprocess specified columns in the existing DataFrame
columns_to_preprocess = ['Headline', 'articleBody']  # specify your columns here

for column in columns_to_preprocess:
    df[column] = df[column].astype(str).apply(preprocess_text)
df['stance_cat'] = df['Stance'].map({'agree':0,'disagree':1,'discuss':2,'unrelated':3}).astype(int)
df['Stance'].value_counts()
# You can save the DataFrame back to the same file if you want to overwrite it
df.to_csv('finaltest.csv', index=False)
```
